```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
data = pd.read_excel("iris.xls")
```

```python
km =KMeans(n_clusters=3,max_iter=200)
km.fit(data.iloc[:,0:3])
km.cluster_centers_
km.n_clusters
```
